```python
import pandas as pd
import numpy as np
```

```python
customer_data_2017 = pd.read_csv("mod2017.csv",index_col = 0)
customer_data_2018 = pd.read_csv("mod2018.csv",index_col = 0)
customer_data_2019 = pd.read_csv("mod2019.csv",index_col = 0)
customer_data = pd.read_csv("customer_data .csv")
```

```python
customer_data_2017.isnull().sum()
```

```python
customer_data_2017.gender = customer_data_2017.gender.fillna("male")
customer_data_2018.gender = customer_data_2018.gender.fillna("male")
customer_data_2019.gender = customer_data_2019.gender.fillna("male")
```

```python
customer_data_2017.birth_year = customer_data_2017.birth_year.fillna(customer_data_2017.birth_year.mean())
customer_data_2018.birth_year = customer_data_2018.birth_year.fillna(customer_data_2018.birth_year.mean())
customer_data_2019.birth_year = customer_data_2019.birth_year.fillna(customer_data_2019.birth_year.mean())
```

```python
customer_data_2017 = customer_data_2017.fillna(0)
customer_data_2018 = customer_data_2018.fillna(0)
customer_data_2019 = customer_data_2019.fillna(0)
```
